```python
totales = df_train.isnull().sum().sort_values(ascending=False)
```

```python
porcentajes = (df_train.isnull().sum() / df_train.isnull().count()).sort_values(ascending=False)
#sumo y lo divido entre lqa cuenta
```

```python
vacias=pd.concat([totales,porcentajes],axis=1,keys=["total","porcen"])
```

```python
vacias[vacias["total"]>0]
#de las vacias en total cuales osn mayor a cero
```

```python
#quiero quitar los que son menos del 5 porciento
df_train=df_train[vacias[vacias["porcen"] < 0.05].index]
```
