```python
import pandas as pd
```

```python
dataframe=pd.read_csv('titanic1.csv')
dataframe.head()
```

```python
train=dataframe.sample(frac=0.8,random_state=200)
#crea un nuevo DataFrame llamado train que consiste en una muestra aleatoria del 80% de las filas del DataFrame original dataframe.
train.head()
```

```python
test=dataframe.drop(train.index)
#crear un conjunto de datos de prueba (test) a partir del DataFrame original (dataframe) eliminando las filas que ya se han seleccionado para el conjunto de 
#entrenamiento (train). Aquí está el análisis detallado:
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#supervivientes y no; y un vector para analizar si en un futuro cuantos podran sobrevivir
#estadistica, algoritmos de predicción 
# se están preparando datos para un modelo de aprendizaje automático.##################
#vamos a crear la matriz   axis=eje 1
X=dataframe.drop('Survived',axis=1)#MATRIZ
#X será una matriz que contiene las características del conjunto de datos, Excluyendo la columna 'Survived'. LA BORRA
#axis=1 indica que se está eliminando una columna en lugar de una fila (por defecto, axis es 0, que se refiere a las filas).
y=dataframe.Survived
#y será un vector que contiene las etiquetas de clase ('Survived') correspondientes a las filas del conjunto de datos.
```

```python
y.head()
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
""" un conjunto de entrenamiento (X_train, y_train) y un conjunto de prueba (X_test, y_test).

los divide en conjuntos de entrenamiento y prueba. Los argumentos son:

X: La matriz de características.
y: El vector de etiquetas.
test_size: El tamaño del conjunto de prueba. En este caso, se establece en 0.3, lo que significa que el 30% 
de los datos se utilizarán para formar el conjunto de prueba, y el 70% restante se utilizará para formar el conjunto de entrenamiento.
"""
```
