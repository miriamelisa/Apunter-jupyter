```python
#Vamos a añadir algo mas para que me salgan las que tienen mas de un determinado porcentaje
round(df.isnull().sum()/len(df)*100,2) > 5
```

```python
columnasvacias = df.columns[round(df.isnull().sum()/len(df)*100,2) > 5]
#Así solo sale el nombre de las que son true, de las que tiene un porcentaje mayor de 5
```

```python
df.columns[round(df.isnull().sum()/len(df)*100,2) > 5]
```

```python
#Quiero borrar estas columnnsa
df.drop(columnasvacias,axis=1,inplace=True)
#(((drop siempre va a borrar estructuras)))
```
